```python
def find_five_digit():
  for d1 in range(1,10):
    for d2 in range(10):
      for d3 in range(10):
        for d4 in range(10):
          for d5 in range(10):
            if d1 !=d2+2: continue
            if d2 !=d3+2: continue
            if d4 !=d3-2: continue
            if d5 !=d4+2: continue
            if(d3+d4+d5) !=d1: continue
            if(d1+d2+d3+d4+d5) !=19: continue
            print(f"The five digit number is {d1}{d2}{d3}{d4}{d5}")
find_five_digit()
```
